Zunaechst bestimmen wir die Waermekapazitaet des Kalorimeters:

Wir fuellen den Kalorimeter mit Wasser bei Raumtemperatur $T_{kalt} = 20^\circ C$, wobei fuer die Temperatur des Kalorimeters diese entspricht:  $T_K = T_{kalt}$

Dann fuegen wir warmes Wasser der Temperatur $T_{warm} = 90.1^\circ C$ hinzu.

Der Betrag der vom warmen Wasser abgegebenen Waerme entspricht der vom kalten Wasser und dem Kalorimeter aufgenommene Waerme.

$$-\Delta Q_{warm} = \Delta Q_K + \Delta Q_{kalt}$$

Fuer die Waermekapazitaet des Kalorimeters ergibt sich

$$C_K = -c_wm_{w, warm}\frac{(T_E-T_{warm})}{(T_E-T_{kalt})}-c_wm_{w, kalt}$$

Die Unsicheheit des Ergebniss wäre dann durch

$$
\Delta C_\mathrm{k} = \sqrt{
\left(\Delta m_\mathrm{w, warm} \left(-c_\mathrm{w} \frac{T_\mathrm{E} - T_{\mathrm{warm}}}{T_\mathrm{E} - T_{\mathrm{kalt}}}\right)\right)^2
+ \left(-\Delta m_\mathrm{w, kalt} c_\mathrm{w}\right)^2
+ \left(\Delta T_\mathrm{E} \left(c_\mathrm{w}m_\mathrm{w, warm}\frac{T_\mathrm{k}}{\left(T_\mathrm{E} - T_\mathrm{k}\right)^2}\right)\right)^2
+ \left(\Delta T_\mathrm{warm} c_\mathrm{w}m_\mathrm{w, warm}\frac{1}{T_\mathrm{E} - T_\mathrm{kalt}}\right)^2
+ \left(\Delta T_\mathrm{kalt} \left(-c_\mathrm{w}m_\mathrm{w, warm}\frac{T_\mathrm{E} - T_\mathrm{warm}}{\left(T_\mathrm{E} - T_\mathrm{kalt}\right)^2}\right)\right)^2}
$$


gegeben.

Wir messen:

**Lösung:**

Um die Temperatur eines Körpers zu erhöhen muss man ihm Energie in Form von Wärme zuführen. Man findet dabei, dass die zugeführte Wärmemenge $\Delta Q$ proportional zur Temperaturänderung $\Delta T$ und der Masse m des Körpers ist:
$$
\Delta Q= c\ m\ \Delta T
$$
Mit c als die spezifische Wärmekapazität.

Im Versuch gilt: $\Delta Q_M =-\Delta Q_W$ aufgrund des Wärmetransports zwischen Metal und Wasser/Kalorimeter. Mit der folgenden Umformung erhällt man dann $c_M$ die Wärmekapazität des Metalles:


$$-\Delta Q_M = \Delta Q_W + \Delta Q_K$$

$$-c_M m_M \Delta T_M = c_W m_W \Delta T_W + C_K \Delta T_K$$

$$c_M = -\frac{(T_E-T_{W})}{(T_E-T_{M})}\frac{(c_Wm_W+C_K)}{m_{M}}$$

wobei $\Delta T_K = \Delta T_{Wasser}$

Die Abweichung ist dann als:

$$
\Delta\,C_M=\sqrt{
\left(\Delta\,m_{M,err} \left( \frac{T_E-T_W}{T_E-T_M}\frac{c_Wm_W+C_K}{m_M^2}\right)\right)^2
+ \left(\Delta\,C_K\cdot\left(-\frac{T_E-T_W}{T_E-T_M}\frac{1}{m_M}\right)\right)^2
+ \left(\Delta\,m_W\cdot\left(-\frac{T_E-T_W}{T_E-T_M}\frac{c_W}{m_M}\right)\right)^2
+ \left(\Delta\,T_{E}\cdot\,\frac{T_M}{\left(T_E-T_M\right)^2}\,\frac{(c_Wm_W+C_K)}{m_{M}} \right)^2
+ \left(\Delta\,T_{W}\cdot\,\frac{(c_Wm_W+C_K)}{m_{M}} \left( \frac{1}{(T_{E} - T_{M})} \right) \right)^2
+ \left(\Delta\,T_{K}\cdot\,\frac{(c_Wm_W+C_K)}{m_{M}} \left( \frac{T_{E} - T_{W}}{(T_{E} - T_{M})^2} \right) \right)^2
}$$
gegeben.

Die neue Werte der Temperatur aus der obigen Abbildung wird nun gefittet.

Dabei erhlten wir folgende Parametern:

|Variablen|Wert|Ungenauigkeit|
|:-:|-|-|
|$\alpha\,\text{(in K)}$|$295.388$|$\pm\,0.0077$|
|$\beta\,\text{(in K)}$|$-205.130$|$\pm\,0.014$|
|$\gamma$|$-19813.6$|$\pm\,3.3$|

Die Qualität der Anpassung ist leider nicht gerade Optimal ($\chi^2\,Wahrscheinlichkeit\,ist\,0$) aber mit diesen Parametern wird jetzt vorerst weitergerechnet.

Die Temperaturänderung $\dot{T}_0(t)$ lässt sich folgendermaßen ableiten:
$$
\begin{matrix}
T_0(t) = \alpha+\beta\cdot e^{\frac{t}{\gamma}}&(1)\\
\rightarrow\dot{T}_0(t) = \frac{\beta}{\gamma}e^{\frac{t}{\gamma}}&(2)\\
\text{Im folgenden Schritt wird die erste Gleichung nach  umgestellt:}\\
t(T) =\gamma\,ln\left(\frac{T-\alpha}{\beta}\right)&(3)\\
\text{Es wird nun die dritte Gleichung in die zweite eingesetzt um so die Änderung in abhängigkeit von der Temperatur zu erhalten:}\\
\dot{T}_0(T)=\frac{T-\alpha}{\gamma}&(4)
\end{matrix}$$
Graphich sieht es folgendermaßen aus:

```python
T_dot_0 = (T-a)/y

plt.plot(T, T_dot_0)
plt.xlabel('T in K')
plt.ylabel('Temp. Änderung in K/s')
plt.title('Wärmegang ohne Heizung')
plt.show()
```

\end{equation}
$$
gegeben. 

$\dot{T}_P(T)$ lässt sich folgendermaßen mit Gleichungen $(4) \text{ und } (8)$ bestimmen:
$$\begin{matrix}
\dot{T}_P(T)=\dot{T}(T)-\dot{T}_0(T)\\
\rightarrow \dot{T}_P(T)=\,ab\left(\frac{T-c}a\right)^{1-\frac 1 b}\,-\,\frac{T-\alpha}{\gamma}
\end{matrix}$$
mit der Masse $m_{Al}=338\,g$ und der Leistung$P_H = U\cdot\,I=10.8V\cdot\,1.85A=19.98$ lässt sich dann folgende Kurve erstellen:

```python
m_kern = 0.338 
P_H= 19.98

T_dot_P = params_messung[0]*params_messung[1]*((T-params_messung[2])/params_messung[0])**(1-1/params_messung[1]) -((T-a)/y)
c_Al= P_H/(m_kern*T_dot_P)




plt.plot(T,c_Al)
plt.xlabel('T in K')
plt.ylabel('c in J/kg*K')
plt.title('Temperaturabhängige Wärmekapazität')
plt.show()
```
