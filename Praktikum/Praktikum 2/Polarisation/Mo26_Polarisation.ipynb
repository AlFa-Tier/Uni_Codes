 ---

von oben: kein Unterschied

von der seite: Helligkeit am rand des glases verändert sich in abhängigkeit des polarisationswinkels

**Lösung:**

Ein mit Wasser gefülltes Glas wird von der Seite mit weißem Licht beschienen. \
Betrachtet man das Wasser von oben und verändert den Winkel des Polarisationsfilters, so gibt es keine sichtbare Änderung in Abhängigkeit des Winkels.

Von der Seite, senkrecht zum einfallenden Licht, kann jedoch beobachtet werden, dass die sichtbare Intensität der Reflexionen am Rand des Glases heller und dunkler werden, je nachdem wie der Winkel des Filters eingestellt ist.

Diese Beobachtung legt die Vermutung nahe, dass das Licht, welches seitlich aus dem Wasser austritt polarisiert ist.

Erklären lässt sich diese Polarisation wie folgt: \
Licht breitet sich im Raum in Form einer elektromagnetischen Welle fort, wobei die elektrischen Felder imm ersenkrecht zur Ausbreitungsrichtung stehen.
Tritt das Licht  in das Wasser ein, so regt es die Wassermoleküle zu Dipolschwingungen an. Analog zu einer Antenne strahlen diese Dipole hauptsächlich senkrecht zur Schwingungsrichtung und kaum bis gar nicht in Schwingungsrichtung. Dies sorgt dafür, dass das zur Seite gebrochene Licht polarisiert sein kann.
