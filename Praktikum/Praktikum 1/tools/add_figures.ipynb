# Einbinden von externen Bildern in Jupyter-notebook 

Das folgende Beispiel zeigt, wie Sie ein externes Bild direkt in Jupyter-notebook einbinden können. Auf diese Weise ist der **direkte Export ins *pdf*-Format** ebenfalls gewährleistet:

```python
# This example assumes that you are running this Jupyter-notebook from the 
# directory: /home/USER/P1-Praktikum/students/tools/
# If this is not the case you should adapt all paths to tools and figures, 
# accordingly. 
from os import getcwd
PWD=getcwd()

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define figure object with predefined size; if not suqare the larger 
# dimension of the image defines the size of the figure.
img = plt.figure("demo", figsize=(10.0, 10.0))
# Use the full filename with absolute path to load the image 
img = mpimg.imread("{PWD}/../figures/IMG_7110.jpg".format(PWD=PWD))
# Do rendering as a background process
imgplot = plt.imshow(img)
# Suppress axes
imgplot.axes.get_xaxis().set_visible(False)
imgplot.axes.get_yaxis().set_visible(False)
# Show image in Jupyter-notebook
plt.show()
```
