# Laser-Optik, Teil A

## Motivation

Mit diesem Versuch lernen Sie die grundlegenden Konzepte von [Lasern](https://de.wikipedia.org/wiki/Laser) und ihrer Anwendung in der modernen Physik kennen. Darüber hinaus können Sie Ihr Verständnis der Eigenschaften von Licht und elektromagnetischen Wellen vertiefen. 

Das Laser (*light amplification by stimulated emission of radiation*)-Prinzip beruht auf der 1916 von Albert Einstein erstmals beschriebenen stimulierten Emission elektromagnetischer Wellen. Der erste funktionsfähige Maser –noch im Mikrowellenbereich– wurde 1954 von [Charles Townes](https://de.wikipedia.org/wiki/Charles_H._Townes) realisiert. Die Übertragung des Maser-Prinzips zu den kürzeren Wellenlängen des sichtbaren Lichts gelang [Theodore Maiman](https://de.wikipedia.org/wiki/Theodore_Maiman) 1960 mit dem ersten [Rubinlaser](https://de.wikipedia.org/wiki/Rubinlaser). Seitdem wurden die Eigenschaften und Leistungsfähigkeit von Lasern kontinuierlich erweitert und verbessert. Laser sind eine brillante Quelle für [hoch-kohärentes Licht](https://de.wikipedia.org/wiki/Koh%C3%A4renzl%C3%A4nge). Sie sind damit Wegbereiter für interferometrische Längenmessungen oder die Untersuchung quantenmechanisch verschränkter Zustände. Sie finden Anwendung beim Nachweis von [Gravitationswellen](https://de.wikipedia.org/wiki/Gravitationswelle) sowie bei der Erzeugung von Qbits im [Quantencomputing](https://de.wikipedia.org/wiki/Quantencomputer). Von der Kommunikation und Datenspeicherung bis hin zur Materialbearbeitung und Medizin sind Laser zu wichtigen Werkzeugen in Wissenschaft und Technik geworden.

**Lösung:**

#### Definition des Brewserwinkel
Fallt ein Lichtstrahl auf eine Grenzfläche zwischen 2 Medien (von dem Medium mit Brechungs-index $n_1$ kommend, in das Medium mit $n_2$ wobei $n_2 > n_1$ gilt), so ist der Brewsterwinkel $\theta$ der Winkel,
fur den der reflektierte Strahl vollständig linear polarisiert ist. Der gebrochene und reflektierte Strahl stehen hierbei senkrecht aufeinander.
Dabei gilt dann:
$$
\tan\left(\theta\right) = \frac{n_2}{n_1}
$$
Um den Brewsterwinkel zu bestimmen gibt es zwei Verfahren:
- Methode der minimalen Reflexion
- Mehtode der maximalen Transmission

Für beide Verfahren wird zuerst eine drehbare Glasscheibe im Weg des Lasers montiert.

#### Methode der minimalen Reflexion
Fällt der Laserstrahl unter dem Brewsterwinkel auf die Glasplatte, wird ein Minimum bei der Reflexion beobachtet, da der Laserstrahl polarisiert ist. Es wird im Idealfall eine sehr geringe Relexion geben und kann daher gut beobachtet werden.

Mit dieser Messung ergab sich zwei Messwerte:
$$
\theta_{Brewster} =60°\; \text{und}\, \approx 58.5°
$$

Der Literatur Wert für den Übergang von Licht von Luft nach Glas lautet $\theta_{literatur} = 57.2°$.

#### Methode der maximalen Transmission
Es wird die Intensität des transmitierten Strahles mithilfe eines SI-Photoelements gemessen und im Form einer Stromstärke $I$ angezeigt. 

#### Diskussion der Messverfahren
Bei dem Messen der Ergebnisse erwieß sich die Methode der minimalen Reflexion als effektiver. 

Das liegt einerseits daran, dass die Änderrung der Transmission minimal war, und somit das Schwanken des angezeigtem Wert größer war als die feinen Zunahmen im Laufe des Versuches. Zudem wird die Mesung sehr stark von der Umgebung (vor allem äußere Lichter oder Schatten) beinflüsst.

Bereits bei dem Diskutieren der beiden Methoden wurde erschlossen, dass dieses Verfahren der Methode der minimalen Reflexion unterlegen ist und wurde daher verworfen.

#### Auswertung und Bestimmung von $n_{Glas}$
Der Brechungsindex für Luft ist: $n_{Luft} \approx 1$
und somit gilt dann für $n_{Glas}$:

$$n_{Glass} = tan\left(\theta_{Brewster}\right)$$

**Lösung:**

#### 2.1 Bestimmung der Breiten der Spalten

##### Aufbau und Durchführung:
Zur Bestuimmung der Spaltbreite $b$ wird der Spalt bei ausreichender Entfernung angebracht damit es vollständig beleuchtet wird. Auf dem Schirm wird dann die Lage der Maxima und Minima gekennzeichnet. 
Für den Versuch werden nur die Minima vorerst verwendet, da diese schärfer zu bestimmen sind.
Danach wird der Abstand zum Nullten Maximum vermessen. Die Abstände werden im Folgenden als $x_{n}$ bezeichnet um die Minima zu unterscheiden.

Folgende Größen sind von relevanz:
- n: Nummer des Minimums
- $L, \sigma_{L}$: Abstand zwischen Spalt und Schirm und seine Unsicherheit $\left(240\,\mathrm{cm}\right)$
- $\lambda$: Die Wellenlänge des Verwendetem Lichtes $\left(632.8\,\mathrm{nm}\right)$
- $x_{n},\sigma_{x}$: Abstand des n-ten Minimums und seine Unsicherheit

##### Auswertung der Messdaten:
Es gilt für die Intensität des gebeugnet Lichtes:

$$
I\left(\Theta\right)\sim \left(
    \frac{\sin\left(\pi\frac b\lambda \sin(\Theta)\right)}{\pi\frac b\lambda\sin(\Theta)}
\right)^2
$$

Unter Verwendung der Kleinwinkelnäherung gilt dann
$$
\frac xL =\tan(\Theta) \approx \Theta = \sin(\Theta)
$$
wobei $x$ der Abstand zum Nullten Minimum ist und $L$ der Abstand zwischen Spalt und Schirm ist.

Um von den gemessenen Abständen zu den verschiedenen Minima zu der Spaltbreite zu gelangen wird folgende Gleichung hergeleitet:
$$
\begin{align}
\sin\left(\pi\frac b\lambda \sin(\Theta)\right) &= 0 &|\arcsin (\,)\\
\frac b\pi \sin(\Theta) &= 1 & | \cdot\left(\frac \lambda b\right)\, &\text{Einsetzen der Kleinwinkelnäherung}\\
\frac xL &= \lambda b & |\cdot(b\cdot \frac Lx)\\
b &= \frac{\lambda L}{x}
\end{align}
$$
Um da verschiedene Ordnungen der Minima verwendet werden wird Gleichung (4) korrigiert zu:

$$
b \cdot x_n= n\,L\,\lambda
$$
Zudem wurde in diesem Schritt eine Umformung der Gleichung gemacht um auf der Form $y = a \cdot x$ zu kommen. Somit kann die Messung von $b$ als Steigung einer Regressionsgerade bestimmt werden.

Es treten die folgende Unsicherheiten auf
$$
\begin{array}{c}
\sigma_{x} = 0.5\, \mathrm{mm} = 0.05\, \mathrm{cm}\\
\sigma_{L} = 0.1\, \mathrm{cm}\\
\sigma_{y} = \sqrt{\left(\frac{\partial y}{\partial L}\cdot \sigma_{L}\right)^2} = \sqrt{\left(n\cdot\lambda\cdot\sigma_{L}\right)^2} = n\lambda\sigma_{L}
\end{array}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
import kafe2
from IPython.display import display, Math 

def y_is (n,L, lam = 632.8E-9):

    y  = (n*L*lam)
    return y

def linear_model(x,b):
    return b*x
```

```python
SP_1_x_n = np.array([-1.9, -1.45, -1.0, -0.5, 0.5, 0.95, 1.45, 1.9]) # in cm
n=np.array([-4, -3, -2, -1, 1, 2, 3, 4])
L:float=  240 # in cm
x_err:float= 0.06 
L_err:float= 0.1
y_err:float = 632.8E-9 * L_err 


SP_1_data = kafe2.XYContainer(x_data=SP_1_x_n, y_data= y_is(n,L))
SP_1_data.label='Spalt 1'
SP_1_data.axis_labels =['Abstand zum $0-ten$ Maximum in $[cm]$','$n\cdot L\cdot\lambda\, in[cm^2]$']
SP_1_data.add_error(axis='x', err_val = x_err)
SP_1_data.add_error(axis='y', err_val = y_err)

SP_1_line_fit=kafe2.Fit(data=SP_1_data,model_function=linear_model)
SP_1_line_fit.do_fit()
SP_1_result = kafe2.Plot([SP_1_line_fit])
SP_1_params = SP_1_line_fit.parameter_values
SP_1_param_errors = SP_1_line_fit.parameter_errors


display(Math(rf'\text{{Die Spaltbreite des ersten Spaltes wird aus dem Fit als }} b_1 = \left({SP_1_params[0]:.02E}\pm{SP_1_param_errors[0]:.02E}\right)\text{{ bestimmt.}} '))
SP_1_result.plot()

plt.show()
```

```python
SP_2_x_n = np.array([-2.1, -1.55, -1.05, -0.5, 0.55, 1.1, 1.6, 2.1])


SP_2_data = kafe2.XYContainer(x_data=SP_2_x_n, y_data= y_is(n,L))
SP_2_data.label='Spalt 2'
SP_2_data.axis_labels =['Abstand zum $0-ten$ Maximum in $[cm]$','$n\cdot L\cdot\lambda\, in[cm^2]$']
SP_2_data.add_error(axis='x', err_val = x_err)
SP_2_data.add_error(axis='y', err_val = y_err)

SP_2_line_fit=kafe2.Fit(data=SP_2_data,model_function=linear_model)
SP_2_line_fit.do_fit()
SP_2_result = kafe2.Plot([SP_2_line_fit])
SP_2_params = SP_2_line_fit.parameter_values
SP_2_param_errors = SP_2_line_fit.parameter_errors

display(Math(rf'\text{{Die Spaltbreite des zweiten Spaltes wird aus dem Fit als }} b_2 = \left({SP_2_params[0]:.02E}\pm{SP_2_param_errors[0]:.02E}\right)\text{{ bestimmt.}}'))
SP_2_result.plot()

plt.show()
```

##### Besprechung der Ergebnisse
Es ist auffällig, dass beide Spalte hier $\approx 0.3\,\mathrm{mm}$ als Spaltbreite hatten obwohl eine der Spalten eine breite von $\approx 0.2\,\mathrm{mm}$ haben sollte.
Anhand der sehr hohen $\chi^2$ Probability des ersten fits, kann erschlossen werden, dass dieser tatsächlich das $\approx 0.3\,\mathrm{mm}$ Spalt ist.
Bei dem zweiten Spalt könnte eine Ungenauigkeit in der Messung dazu geführt haben, dass das Ergebniss so groß ist. Eine weiterer Fehler könnte das Vertauschen des Spaltes sein.
Durch eine Unordnung oder falsches zurücklegen der verschiedenen Spalten könnte es sein das zwie Gitter mit der Spaltbreite $\approx 0.3\,\mathrm{mm}$ Verwendet wurden.

##### Babinetsches Theorem
Das Babinet Theorem besagt, dass die Beugungsmuster fur 2 komplementäre Blenden (Spalt und Steg / Kreisloch und Kreisblende) identisch sind. Dies gilt nur bei der Wellenoptik.

Das Babinet Theorem gilt sowohl in der Fraunhofernäherung als auch für Fresnelbeugung.

##### Aufbau und Erwartung
Es wird hierfür der gleiche Versuchsaufbau verwendet wie in Aufgabe 2.1, lediglich wird aber anstelle der zwei Spalten, einen Spalt mit $b = 0.3 \mathrm{mm}$ und der Steg mit Breite $b= 0.\mathrm{mm}$ verwendet.

Zu erwarten ist, dass die zwei Beugunsmuster größtenteils übereinstimmen. Der Maximum Nullter Ordnung sollte aber unterschiedlich sein, da nach der geometrsichen Optik einen Schatten beim Steg und eien Strich bei der Spalte zu erwarten sind.

##### Ausertung des Versuches
Ein Vergleich der zwei Aufnahmen zeigt, dass die Muster nahezu identisch sind Um den Maximum Nullter Ordnung sind leichte Unterschiede bemerkbar, diese sind Folge zuvor bescprochenen Anteile der Geometrischen Optik.

Zudem ist bei der Scheibe auch eine Abdunklung im bereich des Poisson-Flecks zu erkennen, dieser ist Vermutlich eine Folge des in der geometrischen Optik erwartetem Schattens.

##### Auswertung
Es wird analog zur bestimmung der Spaltbreite gearbeitet. Mit der Mikrometerschraube wird die dicke $d_{schraube} = 70\,\mathrm{\mu m}$. Die durch dem Fit ermittelte dicke des Haars ist $\approx 70\, \mathrm{\mu m}$ und stimmt somit mit dem Wert welcher die Scraube ergab.

```python
def y_is (n,L, lam = 632.8E-9):

    y  = (n*L*lam)
    return y

def linear_model(x,b):
    return b*x



H_x_n=np.array([-4.7, -3.1, -1.5, 1.6, 3.1, 4.8]) # in cm
H_n=np.array([-3, -2, -1, 1, 2, 3])
H_L:float=  174.5 # in cm


H_xy_data = kafe2.XYContainer(x_data=H_x_n, y_data= y_is(H_n,H_L))
H_xy_data.label='Haar von ...'
H_xy_data.axis_labels =['Abstand zum $0-ten$ Maximum in $[cm]$','$n\cdot L\cdot\lambda\, in[cm^2]$']
H_xy_data.add_error(axis='x', err_val = x_err)
H_xy_data.add_error(axis='y', err_val = y_err)

H_line_fit=kafe2.Fit(data=H_xy_data,model_function=linear_model)
H_line_fit.do_fit()
H_result = kafe2.Plot([H_line_fit])
H_params = H_line_fit.parameter_values
H_param_errors = H_line_fit.parameter_errors


display(Math(rf'\text{{Die Haarbreite wird aus dem Fit als }} b_{{Haar}} = \left({H_params[0]:.02E}\pm{H_param_errors[0]:.02E}\right)\text{{ bestimmt.}}'))
H_result.plot()
plt.show()
```

  - Die Hauptmaxima der Einzelspaltbeugung werden breiter und weniger intensiv.
